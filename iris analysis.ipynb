```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Download the iris dataset
iris = sns.load_dataset('iris')

# Define bin sizes as NumPy array
bins = np.array([5, 10, 15, 20])

# Plot histograms for each feature with different bin sizes

plt.figure(figsize=(12, 10))

# Petal Width
plt.subplot(2, 2, 1)
for bin_size in bins:
    plt.hist(iris['petal_width'], bins=bin_size, alpha=0.5, label=f'Bins={bin_size}')

plt.legend()
plt.title('Petal Width Histogram')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')

# Petal Length
plt.subplot(2, 2, 2)
for bin_size in bins:
    plt.hist(iris['petal_length'], bins=bin_size, alpha=0.5, label=f'Bins={bin_size}')

plt.legend()
plt.title('Petal Length Histogram')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')

# Sepal Width
plt.subplot(2, 2, 3)
for bin_size in bins:
    plt.hist(iris['sepal_width'], bins=bin_size, alpha=0.5, label=f'Bins={bin_size}')

plt.legend()
plt.title('Sepal Width Histogram')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')

# Sepal Length
plt.subplot(2, 2, 4)
for bin_size in bins:
    plt.hist(iris['sepal_length'], bins=bin_size, alpha=0.5, label=f'Bins={bin_size}')

plt.legend()
plt.title('Sepal Length Histogram')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset('iris')

# Define the features
features = ['petal_width', 'petal_length', 'sepal_width', 'sepal_length']

# Define the bins
bins = [5, 10, 15, 20]

# Plot histograms for each feature with different bin sizes
for feature in features:
    plt.figure(figsize=(15, 5))
    plt.suptitle(f'{feature.capitalize()} Histogram with Different Bin Sizes', fontsize=16)
    for i, bin_size in enumerate(bins):
        plt.subplot(1, len(bins), i + 1)
        plt.hist(iris[feature], bins=bin_size, alpha=0.5)
        plt.title(f'Bin Size: {bin_size}')
        plt.xlabel(f'{feature.capitalize()}')
        plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset('iris')

# Define the features
features = ['petal_width', 'petal_length', 'sepal_width', 'sepal_length']

# Define the bins
bins = [5, 10, 15, 20]

# Plot histograms for each feature with different bin sizes
for feature in features:
    plt.figure(figsize=(15, 5))
    plt.suptitle(f'{feature.capitalize()} Histogram Comparison', fontsize=16)
    for i, bin_size in enumerate(bins, 1):
        plt.subplot(1, len(bins), i)
        plt.hist(iris[feature], bins=bin_size, alpha=0.5)
        plt.title(f'Bin Size: {bin_size}')
        plt.xlabel(f'{feature.capitalize()}')
        plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```
